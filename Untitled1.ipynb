```python
df1 = df[df['location'].notnull()]
df1 = df1.reset_index()
df1
```

```python

df1[['latitude', 'longtitude']] = df1['location'].str.split(',', n=1, expand=True)
```

```python
m = folium.Map(location=[10.762622,106.660172], tiles='OpenStreetMap', zoom_start=5)
markerCluster = MarkerCluster().add_to(m)
```

```python
for i, row in df1.iterrows():
    lat = df1.at[i,'latitude']
    lng = df1.at[i,'longtitude']
    restaurant = df1.at[i,'area_name']
    popup = df1.at[i,'area_name']
    color = 'blue'
    folium.Marker(location=[lat, lng],popup=restaurant, icon=folium.Icon(color=color)).add_to(markerCluster)
m.save("index.html")
```

```python
def scatterPlot(x,fig):
    plt.subplot(5,2,fig)
    plt.scatter(df_filtered[x],df_filtered['log_price'])
    plt.title(x+' vs logPrice')
    plt.ylabel('logPrice')
    plt.xlabel(x)

def scatter2Plot(x,fig):
    plt.subplot(5,2,fig)
    plt.scatter(df_filtered2[x],df_filtered2['log_price'])
    plt.title(x+' vs logPrice')
    plt.ylabel('logPrice')
    plt.xlabel(x)
plt.figure(figsize=(10,12))

scatterPlot('toilets', 1)
scatterPlot('rooms', 2)
scatterPlot('living_size', 3)
scatter2Plot('length', 4)

plt.tight_layout()
```
